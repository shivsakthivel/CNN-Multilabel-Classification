```python
import tensorflow as tf
import tensorflow.keras as keras
```

```python
train_images = train_generator.flow_from_dataframe(
    dataframe=training,
    x_col='filepaths',
    y_col=['Edema', 'Pleural Effusion'],
    batch_size=32,
    target_size=(512, 512),
    class_mode='raw',
    shuffle = True,
    seed = 42
)

val_images = val_generator.flow_from_dataframe(
    dataframe=validation,
    x_col='filepaths',
    y_col=['Edema', 'Pleural Effusion'],
    batch_size=32,
    target_size=(512, 512),
    class_mode='raw',
    shuffle=True,
    seed=42
)

test_images = test_generator.flow_from_dataframe(
    dataframe=test,
    x_col='filepaths',
    y_col=['Edema', 'Pleural Effusion'],
    batch_size=32,
    target_size=(512, 512),
    class_mode='raw',
    shuffle=False
)
```

```python
# Load the ResNet-50 model with pre-trained weights
resnet50 = tf.keras.applications.ResNet50(weights='imagenet', include_top=False, input_shape=(512, 512, 3))

# Freeze all layers in the pre-trained network
for layer in resnet50.layers:
    layer.trainable = False

# Add a custom fully-connected layer on top of the ResNet-50
x = resnet50.output
x = keras.layers.GlobalAveragePooling2D()(x)
x = keras.layers.Dense(1024, activation='relu')(x)
x = keras.layers.Dropout(0.5)(x)

# Output layer with two binary labels
edema_output = keras.layers.Dense(1, activation='sigmoid', name='edema_output')(x)
pleural_effusion_output = keras.layers.Dense(1, activation='sigmoid', name='pleural_effusion_output')(x)

# Create the final model
model = keras.models.Model(inputs=resnet50.input, outputs=[edema_output, pleural_effusion_output])

# Compile the model
model.compile(optimizer='adam', loss={'edema_output': 'binary_crossentropy', 'pleural_effusion_output': 'binary_crossentropy'}, metrics=['accuracy'])
```

```python
history = model.fit(
    train_images,
    validation_data=val_images,
    epochs=10,
    callbacks=[
        tf.keras.callbacks.EarlyStopping(
            monitor='val_loss',
            patience=10,
            restore_best_weights=True
        )
    ]
)
```
