```python
import tensorflow as tf
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from sklearn.model_selection import train_test_split
```

```python
labels_x = overall_data[["Edema", "Pleural Effusion"]].to_numpy()
labels_x
```

```python
class_counts_x = np.array([np.sum((labels_x == [0, 0]).all(axis=1)), np.sum((labels_x == [0, 1]).all(axis=1)), 
                        np.sum((labels_x == [1, 0]).all(axis=1)), np.sum((labels_x == [1, 1]).all(axis=1))])
class_weights_x = class_counts_x.max() / class_counts_x
class_weights_dict = dict(enumerate(class_weights_x))
```

```python
train_data, test_data = train_test_split(overall_data, test_size=0.2, random_state=42)
val_data, test_data = train_test_split(test_data, test_size=0.5, random_state=42)
```

```python
# Define the ImageDataGenerator
train_gen = ImageDataGenerator(rescale=1./255)
val_gen = ImageDataGenerator(rescale=1./255)
test_gen = ImageDataGenerator(rescale=1./255)
```

```python
# Use flow_from_dataframe method to create the generator
train_generator = train_gen.flow_from_dataframe(
    dataframe=train_data,
    directory=None,
    x_col="Filepath",
    y_col=["Edema", "Pleural Effusion"],
    target_size=(512, 512),
    batch_size=32,
    class_mode="raw",
    shuffle=True
)

val_generator = val_gen.flow_from_dataframe(
    dataframe=val_data,
    directory=None,
    x_col="Filepath",
    y_col=["Edema", "Pleural Effusion"],
    target_size=(512, 512),
    batch_size=32,
    class_mode="raw",
    shuffle=False
)
test_generator = val_gen.flow_from_dataframe(
    dataframe=test_data,
    directory=None,
    x_col="Filepath",
    y_col=["Edema", "Pleural Effusion"],
    target_size=(512, 512),
    batch_size=32,
    class_mode="raw",
    shuffle=False
)
```

```python
# Ensure that we are using GPU
print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))
```

```python
# Create the base ResNet50 model
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(512, 512, 3))
# Add layers on top of the base model to create the final model
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(1024, activation='relu')(x)
x = Dropout(0.5)(x)
# Output layer with two binary labels
edema_output = Dense(1, activation='sigmoid', name='edema_output')(x)
pleural_effusion_output = Dense(1, activation='sigmoid', name='pleural_effusion_output')(x)

model = Model(inputs=base_model.input, outputs=[edema_output, pleural_effusion_output])
```

```python
# Freeze the ResNet50 layers
for layer in model.layers:
    layer.trainable = False
```

```python
# Index at Block 4: 81, Index at Block 5: 143
for layer in model.layers[143:]:
    layer.trainable = True
```

```python
model.summary()
```

```python
model.compile(optimizer= tf.keras.optimizers.Adam(learning_rate=0.0001), loss={'edema_output': 'binary_crossentropy', 'pleural_effusion_output': 'binary_crossentropy'}, metrics=['accuracy'], 
              sample_weight_mode="temporal")
```

```python
history = model.fit(
    train_generator,
    steps_per_epoch=len(train_data)//32,
    epochs=5,
    validation_data=val_generator,
    validation_steps=len(val_data)//32,
    class_weight=class_weights_dict,
    callbacks=[tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)]
)
```
