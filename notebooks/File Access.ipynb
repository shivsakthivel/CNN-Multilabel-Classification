```python
split = pd.read_csv('mimic-cxr-2.0.0-split.csv')
split
```

```python
def construct_filepath(row):
    first_folder = "gs://mimic-cxr-jpg-2.0.0.physionet.org/files"
    subject_id = str(row["subject_id"])
    identifier = "p" + subject_id[:2]
    patient_id = "p" + str(subject_id)
    study_id = "s" + str(row["study_id"])
    filename = row["dicom_id"] + ".jpg"
    return "/".join([first_folder, identifier, patient_id, study_id, filename])
```

```python
paths = split.apply(lambda x: construct_filepath(x), axis = 1)
```

```python
split["filepaths"] = paths
```

```python
training_subset = chexpert[(chexpert["Pleural Effusion"] == 0) | (chexpert["Pleural Effusion"] == 1)]
training_subset
```

```python
training_subset = training_subset[(chexpert["Edema"] == 0) | (chexpert["Edema"] == 1)]
training_subset
```

```python
print ("0 0")
print(len(training_subset[(training_subset['Edema'] == 0) & (training_subset['Pleural Effusion'] == 0)]))
print ("0 1")
print(len(training_subset[(training_subset['Edema'] == 0) & (training_subset['Pleural Effusion'] == 1)]))
print ("1 0")
print(len(training_subset[(training_subset['Edema'] == 1) & (training_subset['Pleural Effusion'] == 0)]))
print ("1 1")
print(len(training_subset[(training_subset['Edema'] == 1) & (training_subset['Pleural Effusion'] == 1)]))
```

There is a definite imbalance in the number of training examples between the consistent and inconsistent classes. It is a potential issue, but on the bright side, there is a good number of cases where both predicted labels are not 1 (19000 out of this subset of 34000), which would otherwise skew the network into only predicting severe cases.

Maybe, after getting a version of the network going, further thought to including more examples can be considered.

```python
new_df = pd.merge(training_subset, split,  how='left', left_on=['subject_id','study_id'], right_on = ['subject_id','study_id'])
new_df
```

```python
pd.unique(metadata['ViewPosition'])
```

```python
non_lateral_films = metadata[(metadata['ViewPosition'].str.contains("PA", na = False)) | (metadata['ViewPosition'].str.contains("AP", na = False)) ]
non_lateral_films
```

```python
final_df = pd.merge(new_df, non_lateral_films,  how='inner', left_on=['subject_id','study_id', 'dicom_id'], right_on = ['subject_id','study_id', 'dicom_id'])
final_df
```

```python
train_dataset = final_df[final_df['split'] == 'train'][["filepaths", "Edema", "Pleural Effusion"]]
validation_dataset = final_df[final_df['split'] == 'validate'][["filepaths", "Edema", "Pleural Effusion"]]
test_dataset = final_df[final_df['split'] == 'test'][["filepaths", "Edema", "Pleural Effusion"]]
```

```python
train_dataset.to_pickle("./train_df.pkl")
validation_dataset.to_pickle("./validation_df.pkl")
test_dataset.to_pickle("./test_df.pkl")
```
